```python
#LOAD VAE

MODEL_FILENAME = 'jun22-dimensional-reduction-beta_1_BEST.h5'

# reparameterization trick
# instead of sampling from Q(z|X), sample eps = N(0,I)
# z = z_mean + sqrt(var)*eps
def sampling(args):
    """Reparameterization trick by sampling fr an isotropic unit Gaussian.
    # Arguments:
        args (tensor): mean and log of variance of Q(z|X)
    # Returns:
        z (tensor): sampled latent vector
    """

    z_mean, z_log_var = args
    batch = K.shape(z_mean)[0]
    dim = K.int_shape(z_mean)[1]
    # by default, random_normal has mean=0 and std=1.0
    epsilon = K.random_normal(shape=(batch, dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon


def load_vae(nodes, beta):
    reg = None
    latent_dim = 2
    in_parameters = Input(shape=(4,), name='in_parameters')
    inputs = in_parameters
    x = Dense(nodes[0], activation='tanh', kernel_regularizer=reg)(inputs)
    x = Dense(nodes[1], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[2], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[3], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[4], activation='tanh', kernel_regularizer=reg)(x)
    z_mean = Dense(latent_dim, name='z_mean', activation='linear', kernel_regularizer=reg)(x)
    z_log_var = Dense(latent_dim, name='z_log_var', activation='linear', kernel_regularizer=reg)(x)

    # use reparameterization trick to push the sampling out as input
    # note that "output_shape" isn't necessary with the TensorFlow backend
    z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

    # instantiate encoder model
    encoder = Model(in_parameters, [z_mean, z_log_var, z], name='encoder')
    #encoder.summary()
    #plot_model(encoder, to_file='vae_mlp_encoder.png', show_shapes=True)

    # build decoder model
    latent_inputs = Input(shape=(latent_dim,), name='z_sampling')
    x = Dense(nodes[4], activation='tanh', kernel_regularizer=reg)(latent_inputs)
    x = Dense(nodes[3], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[2], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[1], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[0], activation='tanh', kernel_regularizer=reg)(x)

    out_parameters = Dense(4, activation='linear')(x)

    # instantiate decoder model
    decoder = Model(latent_inputs, out_parameters, name='decoder')
    #decoder.summary()
    #plot_model(decoder, to_file='vae_mlp_decoder.png', show_shapes=True)

    # instantiate VAE model
    outputs = decoder(encoder(in_parameters)[2])
    vae = Model(
        in_parameters, 
        outputs,
        name='vae_mlp'
    )
    
    vae.compile(optimizer='adam', loss='mean_squared_error', experimental_run_tf_function=False)
    return encoder, vae, decoder


nodes = [100,100,50,25,10]
beta = 0
encoder, vae, decoder = load_vae(nodes, beta)
vae.load_weights(MODEL_FILENAME)
```

```python
#Calculate branching ratio to three leptons

def reset_BR():
    global char_to_e, char_to_mu, char_to_W, char_to_neut_e, char_to_neut_mu, char_to_neut_W 
    global char_to_neut, neut_to_e_e, neut_to_mu_mu, neut_to_Z, neut_to_h0, neut_to_char_e, neut_to_char_mu, neut_to_char_W
    
    #off shell chargino decay to W
    char_to_e = 0
    char_to_mu = 0
    #on shell chargino decay to W
    char_to_W = 0
    #chargino to neutralino
    char_to_neut = 0
    char_to_neut_e = 0
    char_to_neut_mu = 0
    char_to_neut_W = 0
    
    
    #off shell neutralino decay to Z
    neut_to_e_e = 0
    neut_to_mu_mu = 0
    #on shell neutralino decay to Z
    neut_to_Z = 0
    #on shell neutralino decay to h0
    neut_to_h0 = 0
    #neutralino to chargino
    neut_to_char_e = 0
    neut_to_char_mu = 0
    neut_to_char_W = 0 

    
factors_3l = np.zeros(len(BR))

# W Z h0 to 1 and 2 leptons
W_to_l = 0.2164
Z_to_ll = 0.067264
h0_to_ll = 0.01063 + 0.002468 

#off shell chargino decay to lepton
char_to_e = 0
char_to_mu = 0
#on shell chargino decay to W
char_to_W = 0
#chargino to neutralino
char_to_neut_e = 0
char_to_neut_mu = 0
char_to_neut_W = 0
char_to_neut = 0

#off shell neutralino decay to leptons
neut_to_e_e = 0
neut_to_mu_mu = 0
#on shell neutralino decay to Z
neut_to_Z = 0
#on shell neutralino decay to h0
neut_to_h0 = 0
#neutralino to chargino
neut_to_char_e = 0
neut_to_char_mu = 0
neut_to_char_W = 0 

W_mass = 80.379 #GeV
Z_mass = 91.187 #GeV

higgs = np.zeros(len(BR))

for i in range(len(BR)):
    mass = masses[i]
    
    # remove low mass splittings
    if (abs(mass[0] - mass[1]) < 15) or (abs(mass[0] - mass[4]) < 15) or (mass[4] < 103.5):# or (mass[4] - mass[0] < W_mass + 5.0) or (mass[1] - mass[0] < Z_mass + 5.0):
        factors_3l[i] = 0
        continue

        
    BR[i] = BR[i].replace('[','').replace(']','')
    line = BR[i].split(',')
    
    # Loop over and look for "scenario 1" described below
    for j in range(len(line))[::2]:
        # scenario 1, chi02  -> chi01 Z -> l l or chi01 h0 -> ll 
        #             chi+-1 -> chi01 W+- -> l

        #off shell chargino to W
        if line[j] == 'BR(chi^+_1 -> chi^0_1 e^+ nu_e)':
            char_to_e = float(line[j+1])
            # char_to_e = 0
        elif line[j] == 'BR(chi^+_1 -> chi^0_1 mu^+ nu_mu)':
            char_to_mu = float(line[j+1])
            # char_to_mu = 0
        #on shell chargino to W
        elif line[j] == 'BR(chi^+_1 -> chi^0_1 W^+)':
            char_to_W = float(line[j+1])
            # char_to_W = 0
        
        #off shell neutralino to Z ??or h0??
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 e^- e^+)':
            neut_to_e_e = float(line[j+1])
            # neut_to_e_e = 0
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 mu^- mu^+)':
            neut_to_mu_mu = float(line[j+1])
            # neut_to_mu_mu = 0
        #on shell neutralino to Z
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 Z)':
            neut_to_Z = float(line[j+1])
            # neut_to_Z = 0
        #on shell neutralino to h0
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 h^0)':
            neut_to_h0 = float(line[j+1])    
            higgs[i] = neut_to_h0
            # neut_to_h0 = 0

    #Sum all possibilities
    BR1 = (char_to_W * W_to_l) * (neut_to_Z * Z_to_ll) \
        + (char_to_e + char_to_mu) * (neut_to_Z * Z_to_ll) \
        + (char_to_W * W_to_l) * (neut_to_h0 * h0_to_ll) \
        + (char_to_e + char_to_mu) * (neut_to_h0 * h0_to_ll) \
        + (char_to_W * W_to_l) * (neut_to_e_e + neut_to_mu_mu) \
        + (char_to_e + char_to_mu) * (neut_to_e_e + neut_to_mu_mu)

    reset_BR()   
    
    # Loop again, this time looking for "scenario 2" where m(n2) > m(x1)
    for j in range(len(line))[::2]:
        # scenario 2, chi02  -> chi+-1 -> W+- -> l
        #                       W-+    -> l
        #             chi+-1 -> W+- -> l

        #off shell chargino to W
        if line[j] == 'BR(chi^+_1 -> chi^0_1 e^+ nu_e)':
            char_to_e = float(line[j+1]) 
        elif line[j] == 'BR(chi^+_1 -> chi^0_1 mu^+ nu_mu)':
            char_to_mu = float(line[j+1]) 
        #on shell chargino to W
        elif line[j] == 'BR(chi^+_1 -> chi^0_1 W^+)':
            char_to_W = float(line[j+1]) 
        
        #off shell neutralino to chargino W
        elif line[j] == 'BR(chi^0_2 -> chi^+_1 e^- nu_bar_e)' or line[j] == 'BR(chi^0_2 -> chi^-_1 e^+ nu_e)':
            neut_to_char_e += float(line[j+1]) 
        elif line[j] == 'BR(chi^0_2 -> chi^+_1 mu^- nu_bar_mu)' or line[j] == 'BR(chi^0_2 -> chi^-_1 mu^+ nu_mu)':
            neut_to_char_mu += float(line[j+1])
        #on shell neutralino to chargino W
        elif line[j] == 'BR(chi^0_2 -> chi^+_1 W^-)' or line[j] == 'BR(chi^0_2 -> chi^-_1 W^+)':
            neut_to_char_W += float(line[j+1]) 

    #Sum all possibilities
    BR2 = (char_to_W * W_to_l) * (neut_to_char_W * (char_to_W * W_to_l) * W_to_l) \
        + (char_to_e + char_to_mu) * (neut_to_char_W * (char_to_e + char_to_mu) * W_to_l) \
        + (char_to_W * W_to_l) * ((neut_to_char_e + neut_to_char_mu) * (char_to_W * W_to_l)) \
        + (char_to_e + char_to_mu) * ((neut_to_char_e + neut_to_char_mu) * (char_to_e + char_to_mu))   
    
    # Loop again, this time looking for "scenario 3" where m(x1) > m(n2)
    for j in range(len(line))[::2]:
        # scenario 2, chi+-1  -> chi02 -> chi01 Z -> l l or h0 -> l l
        #                        W+-   -> l
        #             chi02 -> any
        
        #off shell chargino to neutralino W
        if line[j] == 'BR(chi^+_1 -> chi^0_2 e^+ nu_e)':
            char_to_neut_e = float(line[j+1]) 
        elif line[j] == 'BR(chi^+_1 -> chi^0_2 mu^+ nu_mu)':
            char_to_neut_mu = float(line[j+1])       
        #on shell chargino to neutralino W ?? this might be unecessary
        elif line[j] == 'BR(chi^+_1 -> chi^0_2 W^+)':
            char_to_neut_W = float(line[j+1]) 
        if 'chi^+_1 -> chi^0_2' in line[j]:
            char_to_neut += float(line[j+1]) 
        
        #off shell neutralino to Z
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 e^- e^+)':
            neut_to_e_e = float(line[j+1])
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 mu^- mu^+)':
            neut_to_mu_mu = float(line[j+1])
        #on shell neutralino to Z
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 Z)':
            neut_to_Z = float(line[j+1])
        #on shell neutralino to h0
        elif line[j] == 'BR(chi^0_2 -> chi^0_1 h^0)':
            neut_to_h0 = float(line[j+1])  
        
    BR3 = (char_to_neut_W * W_to_l) * (neut_to_Z * Z_to_ll) \
        + (char_to_neut_e + char_to_neut_mu) * (neut_to_Z * Z_to_ll) \
        + (char_to_neut_W * W_to_l) * (neut_to_h0 * h0_to_ll) \
        + (char_to_neut_e + char_to_neut_mu) * (neut_to_h0 * h0_to_ll) \
        + (char_to_neut_W * W_to_l) * (neut_to_e_e + neut_to_mu_mu) \
        + (char_to_neut_e + char_to_neut_mu) * (neut_to_e_e + neut_to_mu_mu)

    # chi+-1 -> chi02 -> Z -> ll
    #           W+- -> any 
    # chi02 -> Z -> ll
    BR4 = (char_to_neut) * (neut_to_Z * Z_to_ll) * (neut_to_Z * Z_to_ll)\
        + (char_to_neut) * (neut_to_h0 * h0_to_ll) * (neut_to_h0 * h0_to_ll)\
        + (char_to_neut) * (neut_to_h0 * h0_to_ll) * (neut_to_Z * Z_to_ll)*2\
        + (char_to_neut) * (neut_to_Z * Z_to_ll) * (neut_to_e_e + neut_to_mu_mu)\
        + (char_to_neut) * (neut_to_h0 * h0_to_ll) * (neut_to_e_e + neut_to_mu_mu)\
        + (char_to_neut) * (neut_to_e_e + neut_to_mu_mu)**2

        
    reset_BR()
    
    factors_3l[i] = BR1 + BR2 + BR3 + BR4
               
    BR1 = 0
    BR2 = 0 
    BR3 = 0
    BR4 = 0
```
