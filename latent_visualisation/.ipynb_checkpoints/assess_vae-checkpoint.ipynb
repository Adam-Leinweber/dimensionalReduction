```python
#LOAD TEST DATA WITH XSECS AND BRANCHING RATIOS

fname = '../data/test_data_cross_sections.csv'

params = ['#MSSM11atQ_mA_parameters @MSSM11atQ_mA::primary_parameters::M1', 
          '#MSSM11atQ_mA_parameters @MSSM11atQ_mA::primary_parameters::M2',
          '#MSSM11atQ_mA_parameters @MSSM11atQ_mA::primary_parameters::TanBeta',
          '#MSSM11atQ_mA_parameters @MSSM11atQ_mA::primary_parameters::mu']

with open(fname) as f:
    labels = f.readline().rstrip()
labels = labels.split(',')

N_mass_values = 6
N_values = len(labels) - 2 - N_mass_values # -2 for n2_decays and x1_decays

N_skip = 1 #skip the first entry
N_values -= N_skip

N_lines = 0
with open(fname) as f:
    for line in f.readlines()[1:]:
        N_lines += 1

data = np.empty((N_lines,N_values), float)
BR = np.empty((N_lines), object)
masses = np.empty((N_lines,N_mass_values), float)

N_err = 0
i = 0
with open(fname) as f:
    for line in f.readlines()[1:]:
        line = line.split(',')
        BR_index = 0
        for j in range(len(line)):
            if line[j] == ']':
                N_err += 1
                break
            if 'BR' in line[j]:
                BR_index = j
                break
        if BR_index != N_values+N_skip:
            N_err += 1

        else:
            data[i] = np.array(line[N_skip:N_values+N_skip]).astype(np.float)
            BR[i] = ','.join(line[N_values+N_skip:-N_mass_values])
            masses[i] = np.array(line[-N_mass_values:]).astype(np.float)
            i += 1
data = data[:i,:]
BR = BR[:i]
masses = masses[:i,:]

param_values = data[:,:4].astype(np.float)
param_values[:,[2,3]] = param_values[:,[3,2]] #swap mu and tanbeta

#data = np.concatenate((param_values, masses), axis = 1)
data = param_values
print(data)
```

```python
print(len(data))
for i in range(len(params)):
    print(params[i])
    print(max(data[:,i]))
    print(min(data[:,i]))
    
scaler = joblib.load('MSSMEW_scaler.save')
data = scaler.transform(data)
data[:,1] = np.power(data[:,1],1/2) #unskew M2 value                                                                                                         


reg = None
batch_size = 1000
latent_dim = 2
epochs = 10000

MODEL_FILENAME = 'jun22-dimensional-reduction-beta_1_BEST.h5'

# reparameterization trick
# instead of sampling from Q(z|X), sample eps = N(0,I)
# z = z_mean + sqrt(var)*eps
def sampling(args):
    """Reparameterization trick by sampling fr an isotropic unit Gaussian.
    # Arguments:
        args (tensor): mean and log of variance of Q(z|X)
    # Returns:
        z (tensor): sampled latent vector
    """

    z_mean, z_log_var = args
    batch = K.shape(z_mean)[0]
    dim = K.int_shape(z_mean)[1]
    # by default, random_normal has mean=0 and std=1.0
    epsilon = K.random_normal(shape=(batch, dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon


def kl_loss(true, pred, z_mean, z_log_var):
    loss_parameters = mse(true, pred)

    reconstruction_loss = K.mean(loss_parameters)
    kl_loss = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
    kl_loss = K.sum(kl_loss, axis=-1)
    kl_loss *= -0.5

    return(K.mean((1-beta)*reconstruction_loss + beta * kl_loss))

def wrapper(z_mean, z_log_var):
    def custom_kl_loss(true, pred):
        return kl_loss(true, pred, z_mean, z_log_var)
    return custom_kl_loss


def load_vae(nodes, beta):
    in_parameters = Input(shape=(4,), name='in_parameters')
    # in_parameters = Input(shape=(10,), name='in_parameters')
    inputs = in_parameters
    x = Dense(nodes[0], activation='tanh', kernel_regularizer=reg)(inputs)
    x = Dense(nodes[1], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[2], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[3], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[4], activation='tanh', kernel_regularizer=reg)(x)
    z_mean = Dense(latent_dim, name='z_mean', activation='linear', kernel_regularizer=reg)(x)
    z_log_var = Dense(latent_dim, name='z_log_var', activation='linear', kernel_regularizer=reg)(x)

    # use reparameterization trick to push the sampling out as input
    # note that "output_shape" isn't necessary with the TensorFlow backend
    z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])

    # instantiate encoder model
    encoder = Model(in_parameters, [z_mean, z_log_var, z], name='encoder')
    #encoder.summary()
    #plot_model(encoder, to_file='vae_mlp_encoder.png', show_shapes=True)

    # build decoder model
    latent_inputs = Input(shape=(latent_dim,), name='z_sampling')
    x = Dense(nodes[4], activation='tanh', kernel_regularizer=reg)(latent_inputs)
    x = Dense(nodes[3], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[2], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[1], activation='tanh', kernel_regularizer=reg)(x)
    x = Dense(nodes[0], activation='tanh', kernel_regularizer=reg)(x)

    out_parameters = Dense(4, activation='linear')(x)
    # out_parameters = Dense(10, activation='linear')(x)

    # instantiate decoder model
    decoder = Model(latent_inputs, out_parameters, name='decoder')
    #decoder.summary()
    #plot_model(decoder, to_file='vae_mlp_decoder.png', show_shapes=True)

    # instantiate VAE model
    outputs = decoder(encoder(in_parameters)[2])
    vae = Model(
        in_parameters, 
        outputs,
        name='vae_mlp'
    )

    model_kl_loss = wrapper(z_mean, z_log_var)
    
    vae.compile(optimizer='adam', loss='mean_squared_error', experimental_run_tf_function=False)
    return encoder, vae, decoder
```

```python
nodes = [100,100,50,25,10]
beta = 1e-6
encoder, vae, decoder = load_vae(nodes, beta)
vae.load_weights(MODEL_FILENAME)

def calc_ascore(inputs, outputs):
    return(np.sum((inputs - outputs)**2, axis = 1))
    
ascore = calc_ascore(data, np.array(vae.predict(data, batch_size = 100, verbose = 0)))

plt.hist(ascore, bins = 100)
plt.yscale('log')
plt.show()

a_cut = 0.05
data = data[ascore < a_cut]

print('ratio discarded')
print(len(ascore[ascore > a_cut])/len(ascore))
```
